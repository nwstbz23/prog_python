```python
## Select an appropriate dataset and load in this cell.
## Run the below scripts before we begin
import pandas as pd
import plotly.express as px

## source: https://www.kaggle.com/niharika41298/nutrition-details-for-most-common-foods
nutrients_df = pd.read_csv("https://raw.githubusercontent.com/nwstbz23/prog_python/master/nutrients.csv")
nutrients_df

## Print the first 10 rows of the dataset you will use in the treemap
nutrients_df.head(10)
```

```python
## Plot the Treemap using the selected dataset
## Make sure to use the color and path parameters in .treemap() --> for food nutrients, calories details is shown for the food
fig = px.treemap(nutrients_df, values='Calories', color='Calories', 
                 color_continuous_scale='RdBu', 
                 path=[px.Constant('Category'), 'Food']
)
fig.data[0].hovertemplate = '%{label}<br>Calories info:%{value}' #for text label editing for hover
## For nutrients with zero value, they are replaced as blanks to avoid "Weights sum to zero, can't be normalized" error. 
fig.show()
```

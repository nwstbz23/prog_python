```python
# sample demonstration of the IF, nested IF and ELSE loops
# For this example, float is used to allow keying in of numbers with decimals.
min_score = 0.5
max_score = 100
a = input("Please enter your quiz marks: ")
floatofa = float(a)
if floatofa > max_score:
    print("Your marks", floatofa, "cannot be more than 100")
elif floatofa < min_score:
    print("Your marks", floatofa, "cannot be less than min_score 0.5")
else:
    print("Your total score is", floatofa )
# only numbers must be entered, with decimals and number range allowed.
```

```python
# sample demonstration with WHILE function TRUE and break
# break happens only when the rainbow colours are entered. 
# If not, the message box keeps popping up until a rainbow colour is entered.
while True:
    a = input("Please enter rainbow colours: ").upper()
    if a.isalnum():
        if a.startswith("Violet"):
            print("Good, you entered violet colour")
            break
        elif a.startswith("Indigo"):
            print("Good, you entered indigo colour")
            break
        elif a.startswith("Red"):
            print("Good, you entered red colour")
            break
        elif a.startswith("Blue"):
            print("Good, you entered blue colour")
            break
        elif a.startswith("Yellow"):
            print("Good, you entered yellow colour")
            break
        elif a.startswith("Pink"):
            print("Good, you entered pink colour")
            break
        elif a.startswith("Orange"):
            print("Good, you entered orange colour")
            break
        else:
            print("This is not a rainbow colour")
    else:
        print("This is not alphanumeric number")
# if a special symbol is entered, it will display the third statement of not being an alphanumeric number.
```

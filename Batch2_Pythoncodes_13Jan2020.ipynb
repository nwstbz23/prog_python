```python
# [ ] create, call and test the str_analysis() function for Module 4 lab assignment
def str_analysis():
    while True:
        x = input("enter word or integer: ")
    
        if x.isalpha()==True:
            print(x, "contain all alphabetical characters")
            break
        
        elif x.isdigit()==True:
            x=int(x)
            if x>=99:
                print(x, "is a pretty big number")
                break
            elif x<99:
                print(x,"is a smaller than expected")
                break
        else:
            print(x,"is a suprise! It's neither all alpha nor all digit characters")
            break

str_analysis()
```

```python
# source: https://github.com/rluzuriaga/DEV236x-Introduction-to-Python-Absolute-Beginner/blob/master/Final/Required_FINAL_Project_IntroPy.ipynb
# [ ] create, call and test the adding_report() function for FINAL module exam
def adding_report(user_input = "T"):
    total = 0
    items = ""
    print("Input an integer to add to the total or \"Q\" to quit")
    while True:
        int_input = input("Enter an integer or \"Q\": ")
        if int_input.isdigit():
            if user_input == "T" or user_input == "t":
                total += int(int_input)
            elif user_input == "A" or user_input == "a":
                items += int_input + "\n"
                total += int(int_input)
            else:
                return "OhOhOh 1"
        elif int_input[0] == "Q" or int_input[0] == "q":
            if user_input == "T" or user_input == "t":
                return "\nTotal\n" + str(total)
                break
            elif user_input == "A" or user_input == "a":
                return "\nItems\n" + items + "\nTotal\n" + str(total)
                break
            else:
                return "OhOhOh 2"
                break
        else:
            return int_input + " is an invalid input."
            

user_input = input("Enter \"A\" to print all of the input integers and the total\nOR\nEnter \"T\" to just print the total of all the integers.\n")

if user_input == "A" or user_input == "a" or user_input == "T" or user_input == "t":
    print(adding_report(user_input))
else:
    print("You did not enter \"A\" or \"T\"!! Program will not run.")
```
